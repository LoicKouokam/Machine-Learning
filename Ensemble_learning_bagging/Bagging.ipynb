```python
X=df.drop("Outcome", axis="columns")
y=df.Outcome
```

```python
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()
X_s= scaler.fit_transform(X)
X_s
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X_s, y, stratify=y, random_state=10 )
# stratify: try to maintain the balance of Outcome in the train and test sets
```

```python
y_train.value_counts()
```

```python
201/375
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

scores=cross_val_score(DecisionTreeClassifier(), X, y, cv=5)
scores
```

```python
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(),  
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)

bag_model.fit(X_train, y_train)
bag_model.oob_score_
# n_estimators number of subgroup of dataset , max_samples proportion of original data in each subgroup  
```

```python
bag_model.score(X_test,y_test)
```

```python
scores=cross_val_score(bag_model, X, y, cv=5)
scores.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
scores= cross_val_score(RandomForestClassifier(), X, y, cv=5)
scores.mean() # Sample also feature and  can then performed better
```
